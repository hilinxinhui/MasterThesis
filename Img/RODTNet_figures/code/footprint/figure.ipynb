```python
import numpy as np
import pandas as pd
import scienceplots
import matplotlib.pyplot as plt
# plt.style.use(['science', 'ieee'])
plt.style.use(['science', 'ieee'])
```

```python
new_cols = ['Index', 'Model_Name', 'MSE', 'Time', 'Memory_Usage']
df = pd.read_csv('../../data/performance.csv',names=new_cols, header=0).drop(columns=['Index'])
n_model = len(df)

# 生成不同的颜色
colors = plt.colormaps['tab20'](np.linspace(0, 1, n_model))

# 绘制散点图
# plt.figure(figsize=(10, 8))
scatter = plt.scatter(df['Time'], df['MSE'], s=df['Memory_Usage'] / 1000, c=colors, edgecolors='w')

# 添加模型名称
for i, row in df.iterrows():
    plt.text(row['Time'] + 0.5, row['MSE'], row['Model_Name'], fontsize=8, ha='left')

# 添加标签和标题
plt.xlabel('Inference Time (s)')
plt.ylabel('MSE')
# plt.title('Model Performance Comparison')

# 添加图例（由于模型数量较多，图例可能需要省略或简化）
handles, labels = scatter.legend_elements(prop="sizes", num=3)
# plt.legend(handles, labels, title="Memory Usage (MB)", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.legend(handles, labels, title="Parameter Count")
plt.savefig('./figure_1.png')
```
