```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = {
    'bagging': '../../data/comp_res/12_bagging/task1/RUL_DeNN_bagging_tri_task1_policy1C(4%)-6C_id9_cl300',
    'fusion': '../../data/comp_res/13_fusion/task1/RUL_DeNN_fusion_tri_task1_policy1C(4%)-6C_id9_cl300',
    'voting': '../../data/comp_res/14_voting/task1/RUL_DeNN_voting_tri_task1_policy1C(4%)-6C_id9_cl300',
}
```

```python
for key in d.keys():
    pred = np.load(os.path.join(d[key], 'pred_legacy.npy'))
    true = np.load(os.path.join(d[key], 'true_legacy.npy'))
    assert len(pred) == len(true)
    cl = len(pred)
    pred = pred * cl
    true = true * cl
    np.save(os.path.join(d[key], 'pred.npy'), pred)
    np.save(os.path.join(d[key], 'true.npy'), true)
    plt.plot(pred)
    plt.plot(true)
```
