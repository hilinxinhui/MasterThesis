```python
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'ieee'])
```

```python
# 文件路径和目标sheet配置
file_path = "../../data/p2_敏感性分析.xlsx"
target_sheets = ['batch_size', 'd_model', 'n_layer', 'prob']

# 完整任务名称映射
TASK_MAPPING = {
    'task1 (RUL)': 'MATR',
    'task7 (RUL)': 'Tongji',
    'task8 (RUL)': 'HUST',
    'task9 (RUL)': 'XJTU'
}

COLORS = {
    'MATR': '#1f77b4',    # 蓝色
    'Tongji': '#ff7f0e',  # 橙色
    'HUST': '#2ca02c',    # 绿色
    'XJTU': '#d62728'     # 红色
}

def clean_sensitivity_data(df):
    """增强型数据清洗函数"""
    # 清除metric行（兼容大小写和空格）
    metric_rows = [idx for idx in df.index if 'metric' in str(idx).lower()]
    df = df.drop(metric_rows)
    
    # 去除索引和列名的前后空格
    df.index = df.index.str.strip()
    df.columns = df.columns.astype(str).str.strip()
    
    # 任务名称替换（兼容带括号格式）
    df = df.rename(index=lambda x: TASK_MAPPING.get(x, x))
    
    return df

# 配置显示格式
pd.set_option('display.unicode.east_asian_width', True)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.4f}'.format)

# 创建等间距坐标生成器
def generate_positions(labels):
    return list(range(len(labels)))

# 可视化主函数
def plot_sensitivity(df, param_name, save_path):
    plt.figure(figsize=(3, 3))
    x_labels = [str(v) for v in df.columns]
    x_pos = generate_positions(x_labels)
    
    # 绘制各数据集曲线
    for dataset in df.index:
        clean_name = dataset.split()[0]
        # plt.plot(x_pos, df.loc[dataset], 
        #         marker='o', 
        #         linestyle='-',
        #         linewidth=2,
        #         markersize=8,
        #         color=COLORS[clean_name],
        #         label=clean_name)
        plt.plot(x_pos, df.loc[dataset], marker='s', label=clean_name)
    
    # 坐标轴美化
    plt.xticks(x_pos, x_labels)
    plt.xlabel(f"{param_name}")
    plt.ylabel("RMSE")
    # plt.title(f"Sensitivity Analysis: {param_name}")
    
    # 网格和图例
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend(frameon=True)

    # fig = plt.gcf()
    # fig.set_size_inches(6, 6)  # 强制设置为相同尺寸
    # plt.gca().set_aspect('equal', adjustable='box')
    
    # 保存图像
    plt.tight_layout()
    plt.savefig(f"{save_path}/{param_name}_sensitivity.png", 
               dpi=300, bbox_inches='tight')
    plt.close()

# 遍历处理所有目标sheet
for sheet_name in target_sheets:
    try:
        # 读取原始数据
        raw_df = pd.read_excel(file_path,
                             sheet_name=sheet_name,
                             header=0,
                             index_col=0,
                             engine='openpyxl')
        
        # 执行数据清洗
        processed_df = clean_sensitivity_data(raw_df)

        # 执行可视化
        plot_sensitivity(processed_df, sheet_name, "./")
        print(f"成功生成 {sheet_name} 分析图表")
        
        # 打印处理结果
        print(f"\n{'='*60}")
        print(f"✅ 处理完成: {sheet_name.upper()}")
        print(f"{'-'*60}")
        print(f"参数类型: {processed_df.columns.name or '未命名参数'}")
        print(f"数据维度: {processed_df.shape}")
        print(f"{'-'*60}")
        print(processed_df)
        
    except Exception as e:
        print(f"\n❌ 处理 {sheet_name} 时发生错误: {str(e)}")
        continue

print("\n所有数据处理完成！")
```
