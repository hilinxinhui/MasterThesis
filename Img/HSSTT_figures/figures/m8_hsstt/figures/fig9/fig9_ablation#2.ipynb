```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
```

## TRI dataset

在正文中展示的图片

```python
root_path = '../../results/best_new'
res_paths = glob.glob(os.path.join(root_path, '*tri*'))
for path in res_paths:
    test_id = path.split('/')[-1].split('_')[-1].lower()
    res_files = os.listdir(path)
    assert 'exp_obs.npz' in res_files, f'file not found'
    assert 'exp_obs_re.npz' in res_files, f'file not found'
    res_before = np.load(os.path.join(path, 'exp_obs.npz'))
    res_after = np.load(os.path.join(path, 'exp_obs_re.npz'))
    exp_before, obs_before = res_before['exp'], res_before['obs']
    exp_after, obs_after = res_after['exp'], res_after['obs']

    plt.plot(range(1, 101), range(1, 101), color='black', linestyle='--', label='ideal')
    plt.plot(exp_before, obs_before, label='before calibration')
    plt.plot(exp_after, obs_after, label='after calibration')
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'./tri/tri_b3_{test_id}.png')
    plt.close()
```
